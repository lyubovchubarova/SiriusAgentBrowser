### 3.5 –ö–ª–∞—Å—Å—ã –∏ –º–µ—Ç–æ–¥—ã


````python
class User:
    def __init__(self, name: str, age: int) -> None:
        self.name: str = name
        self.age: int = age
        self.friends: list["User"] = []  # Forward reference –≤ –∫–∞–≤—ã—á–∫–∞—Ö

    def greet(self) -> str:
        return f"Hi, I'm {self.name}"

    def add_friend(self, friend: "User") -> None:
        self.friends.append(friend)

    @classmethod
    def from_dict(cls, data: dict[str, str | int]) -> "User":
        return cls(name=str(data["name"]), age=int(data["age"]))

    @staticmethod
    def validate_age(age: int) -> bool:
        return 0 <= age <= 150


# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
alice = User("Alice", 25)
bob = User.from_dict({"name": "Bob", "age": 30})
alice.add_friend(bob)

print(alice.greet())
print(f"Friends: {[f.name for f in alice.friends]}")
````

### 3.6 Callable (—Ñ—É–Ω–∫—Ü–∏–∏ –∫–∞–∫ –∞—Ä–≥—É–º–µ–Ω—Ç—ã)


````python
from collections.abc import Callable


# –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞—é—â–∞—è –¥—Ä—É–≥—É—é —Ñ—É–Ω–∫—Ü–∏—é
def apply_twice(func: Callable[[int], int], value: int) -> int:
    """–ü—Ä–∏–º–µ–Ω—è–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é –¥–≤–∞–∂–¥—ã.

    Callable[[arg_types...], return_type]
    """
    return func(func(value))


def double(x: int) -> int:
    return x * 2


result = apply_twice(double, 5)  # double(double(5)) = double(10) = 20
print(f"apply_twice(double, 5) = {result}")
````

### 3.7 TypedDict –∏ Protocol


````python
from typing import TypedDict, Protocol


# TypedDict ‚Äî —Å–ª–æ–≤–∞—Ä—å —Å —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä–æ–π
class UserDict(TypedDict):
    name: str
    age: int
    email: str


def process_user(user: UserDict) -> str:
    return f"{user['name']} ({user['age']})"


# Protocol ‚Äî —Å—Ç—Ä—É–∫—Ç—É—Ä–Ω–∞—è —Ç–∏–ø–∏–∑–∞—Ü–∏—è (duck typing)
class Drawable(Protocol):
    def draw(self) -> None: ...


class Circle:
    def draw(self) -> None:
        print("Drawing circle")


class Square:
    def draw(self) -> None:
        print("Drawing square")


def render(shape: Drawable) -> None:
    shape.draw()  # –õ—é–±–æ–π –æ–±—ä–µ–∫—Ç —Å –º–µ—Ç–æ–¥–æ–º draw()


# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
render(Circle())
render(Square())
````

### 4.2 –§—É–Ω–∫—Ü–∏–∏ ‚Äî –¥–µ–ª–∞—é—Ç –æ–¥–Ω—É –≤–µ—â—å


````python
# ‚úÖ –•–û–†–û–®–û: –∫–∞–∂–¥–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–µ–ª–∞–µ—Ç –æ–¥–Ω—É –≤–µ—â—å
from typing import TypedDict as TD


class UserData(TD):
    email: str
    name: str


def validate_user_data(data: dict) -> None:
    """–ü—Ä–æ–≤–µ—Ä—è–µ—Ç –æ–±—è–∑–∞—Ç–µ–ª—å–Ω—ã–µ –ø–æ–ª—è."""
    required = ["email", "name"]
    for field in required:
        if not data.get(field):
            raise ValueError(f"{field} required")


def normalize_user_data(data: dict) -> UserData:
    """–ù–æ—Ä–º–∞–ª–∏–∑—É–µ—Ç –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è."""
    return UserData(
        email=data["email"].lower().strip(), name=data["name"].strip().title()
    )


def process_user_data(user_data: dict) -> UserData:
    """–û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è."""
    validate_user_data(user_data)
    return normalize_user_data(user_data)


# –¢–µ—Å—Ç
test_data = {"email": "  ALICE@Example.COM  ", "name": "alice smith"}
result = process_user_data(test_data)
print(f"Processed: {result}")
````

### 4.3 Docstrings


````python
def calculate_discount(
    price: float, discount_percent: float, min_price: float = 0.0
) -> float:
    """–†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ü–µ–Ω—É —Å–æ —Å–∫–∏–¥–∫–æ–π.

    Args:
        price: –ò—Å—Ö–æ–¥–Ω–∞—è —Ü–µ–Ω–∞ —Ç–æ–≤–∞—Ä–∞.
        discount_percent: –ü—Ä–æ—Ü–µ–Ω—Ç —Å–∫–∏–¥–∫–∏ (0-100).
        min_price: –ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ü–µ–Ω–∞ (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 0).

    Returns:
        –¶–µ–Ω–∞ –ø–æ—Å–ª–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è —Å–∫–∏–¥–∫–∏.

    Raises:
        ValueError: –ï—Å–ª–∏ –ø—Ä–æ—Ü–µ–Ω—Ç —Å–∫–∏–¥–∫–∏ –≤–Ω–µ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ 0-100.

    Example:
        >>> calculate_discount(100, 20)
        80.0
    """
    if not 0 <= discount_percent <= 100:
        raise ValueError("Discount must be between 0 and 100")

    discounted = price * (1 - discount_percent / 100)
    return max(discounted, min_price)


# –¢–µ—Å—Ç
print(calculate_discount(100, 20))  # 80.0
print(calculate_discount(50, 90, min_price=10))  # 10.0
````
